```python
from langchain_community.retrievers import (
    PineconeHybridSearchRetriever,
)
from langchain_community.embeddings import HuggingFaceEmbeddings
import sys
import os
from pinecone import Pinecone

embed_model = HuggingFaceEmbeddings()

api_key = os.getenv('PINECONE_API_KEY')
```

```python
pc = Pinecone(api_key)

INDEX_NAME = "data-management-example"
```

# Hybrid Search

Hybrid Search is a combination of having both a keyword style search and a vector style search (Keyword lookup + Semantic lookup). One of the examples:
1. BM25 

## BM 25

BM25 is a ranking function that ranks a set of documents based on the query terms appearing in each document, regardless of the inter-relationship between the query terms within a document (e.g., their relative proximity). It is not a single function, but actually a whole family of scoring functions, with slightly different components and parameters. It is used by search engines to rank matching documents according to their relevance to a given search query and is often referred to as “Okapi BM25,” since the Okapi information retrieval system was the first system implementing this function [1]. 

Advantages:
1. Very quick to compute

```python
index = pc.Index(INDEX_NAME)
```

```python
from pinecone_text.sparse import BM25Encoder

bm25_encoder = BM25Encoder().default()

retriever = PineconeHybridSearchRetriever(
    embeddings=embed_model, sparse_encoder=bm25_encoder, index=index
)

```

```python
retriever.invoke('economics')
```
